# Unemployment Analysis during Covid 19


```python
import numpy as np  
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt  
import plotly.express as pex
data = pd.read_csv("https://raw.githubusercontent.com/AritraPal13/Covid-19-Unemployment-Analysis/main/CovidUnemployment.csv")
data
```

```python
data.describe(include="all")
```

```python
data.mean()
```

```python
h1= data.isnull().sum()           #Using isnull() to select all rows with NaN under a single DataFrame column:
h1  
```

```python
data.columns= ["States","Date","Estimated Unemployment Rate","Estimated Employed", "Estimated Labour Participation Rate","Region","longitude","latitude"]
```

## Analysis 1:

```python
plt.figure(figsize=(15,10))
sns.histplot(x="Estimated Unemployment Rate",hue="Region",data=data).set(title="Unemployment rate according to Region")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.displot(data, x="Estimated Unemployment Rate",hue="Region", kind="kde",height=8).set(title="Unemployment Rate according to Region")
plt.grid(True)
plt.show
```

## Analysis 2:

```python
plt.figure(figsize=(15,10))
sns.histplot(x="Estimated Employed",hue="Region",data=data).set(title="Estimated No of Employee according to Region")
plt.grid(True)
plt.show()
```

```python
sns.displot(data, x="Estimated Employed",hue="Region", kind="kde",height=8).set(title="Estimated No of Employee according to Regions")
plt.grid(True)
plt.show
```
